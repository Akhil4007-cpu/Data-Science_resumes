```python
import zipfile

with zipfile.ZipFile('../data/raw/retail-orders.zip', 'r') as zip_ref:
    zip_ref.extractall('../retail-orders')
```

```python
#read the data
import pandas as pd
df = pd.read_csv('../retail-orders/orders.csv', na_values=['Not Available', 'unknown'])
```

```python
df.head(20)
```

```python
df['Ship Mode'].unique()
```

```python
df.isna().sum()
```

```python
#rename the column - lower case and underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head(5)
```

```python
#check columns and see if derived columns can be created
df['discount'] = df['list_price']*df['discount_percent']*.01
```

```python
df.head(5)
```

```python
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head(5)
```

```python
#convert date datatype
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
```

```python
df.dtypes
```

```python
#drop columns - now not needed
df.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace=True)
```

```python
df.head(5)
```
